```python
import numpy as np 
import matplotlib.pyplot as plt 
import cv2
```

```python
img1 = np.zeros(shape=(500,500,3), dtype = np.int32)
```

```python
img1.shape
```

```python
plt.imshow(img1)
```

```python
cv2.rectangle(img1,pt1=(300,50),pt2=(400,100),color=(255,0,0),thickness=2)
plt.imshow(img1)
```

```python
cv2.rectangle(img1,pt1=(250,250),pt2=(450,450),color=(0,255,0),thickness=3)
plt.imshow(img1)
```

```python
cv2.circle(img1,center=(50,50),radius=25,color=(0,0,255),thickness =3)
plt.imshow(img1)
```

```python
cv2.circle(img1,center=(150,150),radius=35,color=(150,0,255),thickness =-1)
plt.imshow(img1)
```

```python
cv2.line(img1,pt1=(500,500),pt2=(0,0),color=(255,255,255),thickness=3)
plt.imshow(img1)
```

```python
cv2.line(img1,pt1=(50,100),pt2=(150,250),color=(200,100,50),thickness=3)
plt.imshow(img1)
```
