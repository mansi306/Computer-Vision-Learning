```python
import numpy as np 
import matplotlib.pyplot as plt 
import cv2
```

```python
img2 = np.zeros(shape=(500,500,3),dtype = np.int32)
plt.imshow(img2)
```

```python
font = cv2.FONT_HERSHEY_DUPLEX
cv2.putText(img2,text="MANSI patil",org=(10,450),fontFace=font,fontScale=2,color=(255,255,255),thickness=2)
plt.imshow(img2)
```

```python
font = cv2.FONT_ITALIC
cv2.putText(img2,text="MANSI Patil",org=(10,250),fontFace=font,fontScale=2,color=(255,0,0),thickness=2)
plt.imshow(img2)
```

```python
img3 = np.zeros(shape=(500,500,3),dtype=np.int32)
plt.imshow(img3)
```

```python
vertices = np.array([[50,50],[400,50],[400,300],[100,300]],np.int32)
```

```python
cv2.polylines(img3,[vertices],isClosed=True,color=(255,0,0),thickness=3)
plt.imshow(img3)
```

```python
vertices1 = np.array([[100,400],[400,400],[250,200]],np.int32)
```

```python
cv2.polylines(img3,[vertices1],isClosed=True,color=(255,255,255),thickness=3)
plt.imshow(img3)
```

```python
vertices2 = np.array([[50,400],[50,200],[250,50],[450,200],[450,400]],np.int32)
```

```python
cv2.polylines(img3,[vertices2],isClosed=True,color=(0,255,0),thickness=3)
plt.imshow(img3)
```

```python
cv2.circle(img3,center=(250,200),radius=25,color=(0,0,255),thickness=-1)
plt.imshow(img3)
```

```python
vertices3 = np.array([[50,400],[50,200],[250,50],[450,200],[450,400],[250,450]],np.int32)
```

```python
cv2.polylines(img3,[vertices3],isClosed=True,color=(150,255,200),thickness=3)
plt.imshow(img3)
```
